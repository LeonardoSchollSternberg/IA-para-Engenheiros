```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados_pinguins = pd.read_csv("./tamanho_pinguins.csv")
```

```python
dados_pinguins = dados_pinguins[['species','culmen_length_mm','sex']]
```

```python
dados_pinguins.head()
```

```python
dados_pinguins.shape
```

```python
dados_pinguins.dtypes
```

```python
sns.barplot(data = dados_pinguins, x= dados_pinguins['species'], y=dados_pinguins['culmen_length_mm'],estimator = np.median)
```

```python
plt.figure(figsize= (12,6))

ax = sns.barplot(data = dados_pinguins,x=dados_pinguins['species'],y=dados_pinguins['culmen_length_mm'],estimator = np.median)
ax.set_xlabel('Espécies',fontsize = 15)
ax.set_ylabel('Comprimento do Cúlmen (mm)', fontsize = 15)
ax.set_title('Análise Bivariada do Comprimento do Cúlmen e Espécies', fontsize= 20)
```

```python
sns.boxplot( data = dados_pinguins, x= dados_pinguins['species'] ,  y= dados_pinguins["culmen_length_mm"])
```

```python
plt.figure(figsize= (12,6))
ax = sns.boxplot(data = dados_pinguins, x= dados_pinguins['species'], y= dados_pinguins["culmen_length_mm"])
ax.set_xlabel('Espécies',fontsize = 15)
ax.set_ylabel('Comprimento do Cúlmen em mm', fontsize = 15)
ax.set_title('Análise Bivariada do Comprimento do Cúlmen e das Espécies', fontsize= 20)
```

```python
dados_pinguins_machos = dados_pinguins.loc[dados_pinguins['sex']== 'MALE',:]
dados_pinguins_femeas = dados_pinguins.loc[dados_pinguins['sex']== 'FEMALE',:]
sns.histplot( data = dados_pinguins_machos, x= dados_pinguins_machos["culmen_length_mm"], alpha=0.5, color = 'blue')
sns.histplot( data = dados_pinguins_femeas, x= dados_pinguins_femeas["culmen_length_mm"], alpha=0.5, color = 'red')
```

```python
plt.figure(figsize= (12,6))
ax = sns.histplot( data = dados_pinguins_machos, x= dados_pinguins_machos["culmen_length_mm"], alpha=0.5, color = 'blue')
ax = sns.histplot( data = dados_pinguins_femeas, x= dados_pinguins_femeas["culmen_length_mm"], alpha=0.5, color = 'red')
ax.set_xlabel('Comprimento do Cúlmen em mm',fontsize = 15)
ax.set_ylabel('Contagem de Registros', fontsize = 15)
ax.set_title('Análise Bivariada do Comprimento do Cúlmen e Sexo', fontsize= 20)
```
