```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('C:/Users/abalb/Desktop/Dados_Engenheiro.csv')
df
```

```
df.describe()
```

```
df = pd.get_dummies(df,drop_first = True)
df
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_normalizado = scaler.fit(df)
df_normalizado = scaler.transform(df)
```

```
from sklearn.neighbors import LocalOutlierFactor
clf = LocalOutlierFactor(n_neighbors = 40, contamination = .1, metric = 'euclidean')
clf.fit(df_normalizado)
```

```
df_normalizado=pd.DataFrame(df_normalizado,columns=['Petroleum_Engineer_Age',
                                          'Petroleum_Engineer_Income (K$)','Spending_Habits (From 1 to 100)',
                                          'Petroleum_Engineer_Gender_Male'])
df_normalizado['Scores'] = clf.negative_outlier_factor_
plt.figure(figsize=(12,8))
plt.hist(df_normalizado['Scores'])
plt.title('Histograma do de Outlier Negativo')
```

```
df_normalizado['Anomaly'] = clf.fit_predict(df_normalizado.iloc[:,:4])
Anomalia = df_normalizado.loc[df_normalizado['Anomaly']==-1]
Anomalia_index = list(Anomalia.index)
Anomalia
```

```
plt.figure(figsize = (12,8))
groups = df_normalizado.groupby("Anomaly")
for name, group in groups:
    plt.plot(group['Petroleum_Engineer_Income (K$)'], group['Spending_Habits (From 1 to 100)'], 
             marker = "o", linestyle='',label=name)
plt.xlabel('Renda do Engenheiro')
plt.ylabel('Hábitos de Consumo do Engenheiro')
plt.title('LOF identificando as Anomalias')
```
