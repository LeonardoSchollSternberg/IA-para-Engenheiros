```python
import numpy as np
```

```python
Nomes_dos_Alunos = ['AA', 'AB', 'AC', 'AD', 'AE']
Notas_IC01 = [8, 5, 6, 7, 6]
Notas_IC02 = [9, 8, 5, 5, 6]
Notas_IC03 = [6, 9, 5, 9, 10]
```

```python
Media_Notas = np.zeros(5)
print(Media_Notas)
#
for i, nome in enumerate(Nomes_dos_Alunos):
    Media_Notas[i] = np.mean([Notas_IC01[i],Notas_IC02[i],Notas_IC03[i]])
#
print(Media_Notas)
```

```python
# relatório em um formato mais bonito
for i, nome in enumerate(Nomes_dos_Alunos):
    print("Média para {} : {}".format(nome,Media_Notas[i]))
```
