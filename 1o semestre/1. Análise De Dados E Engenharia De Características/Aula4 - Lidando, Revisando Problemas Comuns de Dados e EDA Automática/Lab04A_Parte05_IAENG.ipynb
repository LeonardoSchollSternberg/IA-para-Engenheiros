# importar o algoritmo Floresta de Isolamento da biblioteca scikit-learn

```
from sklearn.ensemble import IsolationForest
clf = IsolationForest(n_estimators = 100, max_samples = 250,random_state = 100, contamination = 0.1)
clf.fit(df)
```

```
df['Scores'] = clf.decision_function(df)
plt.figure(figsize = (12,8))
plt.hist(df['Scores'])
plt.title('Histograma do Escore Médio de Anomalias: Escore Negativo = amostras anômalas')
```

```
df['Anomaly'] = clf.predict(df.iloc[:,:4])
Anomaly=df.loc[df['Anomaly']==-1]
Anomaly_index=list(Anomaly.index)
Anomaly
```

```
plt.figure(figsize = (12,8))
groups = df.groupby('Anomaly')
for name, group in groups:
    plt.plot(group['Petroleum_Engineer_Income (K$)'],
             group['Spending_Habits (From 1 to 100)'],marker='o', linestyle='', label=name)
plt.xlabel('Renda do Engenheiro de Petróleo')
plt.ylabel('Hábitos de Consumo')
plt.title('Floresta Isolada - Anomalias')
```
