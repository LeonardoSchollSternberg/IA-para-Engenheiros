```
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
wine_df = pd.read_csv("C:/Users/abalb/Desktop/vinho.csv")
print(wine_df.head())
```

```
plt.scatter(wine_df.values[:,0], wine_df.values[:,1])
plt.title("Base de Dados sobre Vinhos")
plt.xlabel("Leitura do OD: Denominação de Origem")
plt.ylabel("Prolina")
plt.show()
```

```
km = KMeans(3)
km_clusters = km.fit_predict(wine_df) 
ac = AgglomerativeClustering(3, linkage='average')
ac_clusters = ac.fit_predict(wine_df)
```

```
db_param_options = [[20,5],[25,5],[30,5],[25,7],[35,7],[35,3]]
for ep,min_sample in db_param_options:
    db = DBSCAN(eps=ep, min_samples = min_sample)
    db_clusters = db.fit_predict(wine_df)
    print("Eps: ", ep, "Min Samples: ", min_sample)
    print("Agrupamento com o DBSCAN: ", silhouette_score(wine_df, db_clusters))
```

```
db = DBSCAN(eps=35, min_samples = 3)
db_clusters = db.fit_predict(wine_df)
```

```
db_clusters
```

```
plt.title("Clusters usando o K-Means")
plt.scatter(wine_df['OD_read'], wine_df['Proline'], c=km_clusters,s=50, cmap='tab20b')
plt.show()
plt.title("Clusters usando o Agrupamento Aglomerativo")
plt.scatter(wine_df['OD_read'], wine_df['Proline'], c=ac_clusters,s=50, cmap='tab20b')
plt.show()
plt.title("Clusters usando o DBSCAN")
plt.scatter(wine_df['OD_read'], wine_df['Proline'], c=db_clusters,s=50, cmap='tab20b')
plt.show()
```

```
print("Escore da Silhueta para a Famosa Base de Dados sobre Vinhos:\n")
print("Agrupamento K-Means: ", silhouette_score(wine_df, km_clusters))
print("Agrupamento Aglomerativo: ", silhouette_score(wine_df, ac_clusters))
print("Agrupamento DBSCAN: ", silhouette_score(wine_df, db_clusters))
```
