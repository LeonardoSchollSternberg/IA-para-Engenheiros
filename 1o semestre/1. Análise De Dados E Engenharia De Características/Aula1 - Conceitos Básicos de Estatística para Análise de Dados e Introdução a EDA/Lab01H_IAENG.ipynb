```python
import pandas as pd
import numpy as np
```

```python
base_adulto = pd.read_csv('./adulto.csv')
```

```python
base_adulto
```

```python
base_adulto.describe()
```

```python
def MultiplicarPOR2(n):
    return n*2
```

```python
base_adulto.age.apply(MultiplicarPOR2)
```

```python
total_atributo_fnlwgt = base_adulto.fnlwgt.sum()
```

```python
def CalculatarPercentagem(v):
    return v/total_atributo_fnlwgt*100
```

```python
base_adulto.fnlwgt = base_adulto.fnlwgt.apply(CalculatarPercentagem)
```

```python
base_adulto
```

```python
total_atributo_fnlwgt = base_adulto.fnlwgt.sum()
```

```python
base_adulto.fnlwgt = base_adulto.fnlwgt.apply(lambda v: v/total_atributo_fnlwgt*100)
```
