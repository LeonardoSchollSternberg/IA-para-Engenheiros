```python
base_adulto = pd.read_csv('./adulto.csv')
```

```python
mascara_booleana = base_adulto.education == 'Preschool'
print('Média: {}'.format(np.mean(base_adulto[mascara_booleana].age)))
print('Mediana: {}'.format(np.median(base_adulto[mascara_booleana].age)))
```

```python
mascara_booleana1 = base_adulto['education-num'] > 10
mascara_booleana2 = base_adulto['education-num'] < 10
#
print('Mais do que 10 anos de educação - Ganho de Capital: {}'
      .format(np.mean(base_adulto[mascara_booleana1].capitalGain)))
print('Menos do que 10 anos de educação - Ganho de Capital: {}'
      .format(np.mean(base_adulto[mascara_booleana2].capitalGain)))
```
