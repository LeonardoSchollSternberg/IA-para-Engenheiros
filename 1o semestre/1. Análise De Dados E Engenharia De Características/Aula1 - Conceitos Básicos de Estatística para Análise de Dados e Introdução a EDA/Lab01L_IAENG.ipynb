```python
import pandas as pd
```

```python
precos_casas = pd.read_csv("./PrecosCasas.csv")
```

```python
precos_casas
```

```python
precos_casas = precos_casas[['Zip', 'Price', 'Area', 'Room']]
```

```python
precos_casas.head()
```

```python
precos_casas.shape
```

```python
precos_casas.dtypes
```

```python
precos_casas['Preco/m^2'] = precos_casas['Price']/precos_casas['Area']
```

```python
precos_casas.head()
```

```python
# Visualizando Dados no Matlplotlib
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
classificar_precos_casas = precos_casas.sort_values('Price', ascending = False)
```

```python
classificar_precos_casas.head()
```

```python
plt.figure(figsize= (12,6))
x = classificar_precos_casas['Zip'][0:10]
y = classificar_precos_casas['Price'][0:10]
plt.bar(x,y)
plt.show()
```

```python
plt.figure(figsize= (12,6))
plt.bar(x,y)
plt.title('Top 10 Áreas com os preços de casas mais altos', fontsize=15)
plt.xlabel('Código Postal', fontsize = 12)
plt.xticks(fontsize=10)
plt.ylabel('Preços das Casas em Milhões', fontsize=12)
plt.yticks(fontsize=10)
plt.show()
```

```python
# Subplots
fig, ax = plt.subplots(figsize=(40,18))
x = classificar_precos_casas['Zip'][0:10]
y = classificar_precos_casas['Price'][0:10]
y1 = classificar_precos_casas['Preco/m^2'][0:10]
plt.subplot(1,2,1)
plt.bar(x,y)
plt.xticks(fontsize=17)
plt.ylabel('Preços em milhões', fontsize=25)
plt.yticks(fontsize=20)
plt.title('Top 10 Áreas com os preços de casas mais altos', fontsize=25)
plt.subplot(1,2,2)
plt.bar(x,y1)
plt.xticks(fontsize=17)
plt.ylabel('Preços Casas por metro quadrado', fontsize=25)
plt.yticks(fontsize=20)
plt.title('Top 10 Áreas com os preços mais altos por metro quadrado', fontsize=25)
plt.show()
```
