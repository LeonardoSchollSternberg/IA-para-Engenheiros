```
import pandas as pd
base = pd.read_csv("C:/Users/abalb/Desktop/diabetesgravidez.csv")
base.head()
```

```
conj_caract = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
caract = base[conj_caract] 
alvo = base.label
from sklearn.model_selection import train_test_split
caract_trein, caract_teste, alvo_trein, alvo_teste = train_test_split(caract, alvo, test_size=0.3, random_state=1)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
logreg = LogisticRegression(solver='lbfgs')
logreg.fit(caract_trein,alvo_trein)
predicoes = logreg.predict(caract_teste)
print("Logistic Regression Model Accuracy:",accuracy_score(alvo_teste, predicoes))
```
