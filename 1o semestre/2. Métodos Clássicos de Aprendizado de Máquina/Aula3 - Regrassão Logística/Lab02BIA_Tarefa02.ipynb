```
# VER Nota de Aula sobre o OLS
```

```
from scipy.optimize import fmin
x = np.array([2.2, 4.3, 5.1, 5.8, 6.4, 8.0])
y = np.array([0.4, 10.1, 14.0, 10.9, 15.4, 18.5])
SQE = lambda a, x, y: np.sum((a[0] + a[1]*x - y) ** 2)
a0,a1 = fmin(SQE, [0,1], args=(x,y)); 
plt.plot(x, y, 'ro')
plt.plot([0,10], [a0, a0+a1*10], alpha=0.8) 
for xi, yi in zip(x,y):
    plt.plot([xi]*2, [yi, a0+a1*xi], "k:") 
plt.xlim(2, 9); plt.ylim(0, 20) 
plt.savefig("C:/Users/abalb/Desktop/ExemplosErros.png", dpi = 300, bbox_inches = 'tight')
```

```
# o SAE
```

```
x = np.array([2.2, 4.3, 5.1, 5.8, 6.4, 8.0])
y = np.array([0.4, 10.1, 14.0, 10.9, 15.4, 18.5])
SAE = lambda b, x, y: np.sum(abs(b[0] + b[1]*x - y)) 
b0,b1 = fmin(SAE, [0,1], args=(x,y));
plt.plot(x, y, 'ro')
plt.plot([0,10], [a0, a0+a1*10], alpha=0.8) 
plt.plot([0,10], [b0, b0+b1*10], alpha=0.8) 
for xi, yi in zip(x,y):
    plt.plot([xi]*2, [yi, a0+a1*xi], "k:") 
plt.xlim(2, 9); plt.ylim(0, 20) 
plt.savefig("C:/Users/abalb/Desktop/ExemplosErros_SAE_e_SQE.png", dpi = 300, bbox_inches = 'tight')
```
