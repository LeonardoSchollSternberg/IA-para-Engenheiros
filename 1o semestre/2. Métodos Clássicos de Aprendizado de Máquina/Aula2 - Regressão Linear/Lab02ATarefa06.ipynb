```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
dados = sns.load_dataset('tips')
```

```python
size = dados["size"]
size.loc[:15] = np.nan
size.head(20)
```

```python
size.shape
```

```python
size.isnull().sum()
```

```python
mean = size.mean()
mean = round(mean)
print(mean)
```

```python
size.fillna(mean, inplace=True)
size.head(20)
```

```python
plt.hist(size)
plt.show()
```

```python
min_val = size.mean() - (3 * size.std())
print(min_val)
```

```python
max_val = size.mean() + (3 * size.std())
print(max_val)
```

```python
outliers = size[size > max_val]
outliers.count()
```

```python
print(outliers)
```

```python
age = size[size <= max_val]
age.shape
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
enc = LabelEncoder()
dados["sex"] = enc.fit_transform(dados['sex'].astype('str'))
dados["smoker"] = enc.fit_transform(dados['smoker'].astype('str'))
dados["day"] = enc.fit_transform(dados['day'].astype('str'))
dados["time"] = enc.fit_transform(dados['time'].astype('str'))
```

```python
dados.head()
```

```python
# AQUI COMEÇA A ATIVIDADE 08
```

```python
# Usando a variável dados, que contém todo o conjunto de dados, normalize o
# dados usando a fórmula de normalização e armazene-os em uma nova variável chamada
# dados_normalizadas. Imprima os 10 principais valores:
```

```python
dados_normalizados = (dados - dados.min())/(dados.max()-dados.min())
dados_normalizados.head(10)
```

```python
# agora padronizando os dados
```

```python
dados_padronizados = (dados - dados.mean())/dados.std()
dados_padronizados.head(10)
```
