```python
import pandas as pd
dados = pd.read_csv('./publicidade.csv')
dados.head()
```

```python
X = dados[['TV', 'Radio', 'Jornal']] 
y = dados.Vendas
```

```python
from sklearn.model_selection import train_test_split
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg_lin = LinearRegression()
reg_lin.fit(X_treinamento, y_treinamento)
predicoes = reg_lin.predict(X_teste)
print(reg_lin.intercept_)
print(reg_lin.coef_)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print('Erro Absoluto Médio(MAE):', mean_absolute_error(y_teste,predicoes))  
print("Erro Quadrático Médio(MSE):", mean_squared_error(y_teste, predicoes))  
print("Raiz do Erro Quadrático Médio(RMSE):", np.sqrt(mean_squared_error(y_teste, predicoes)))
print("Escore R2-Quadrado:",r2_score(y_teste, predicoes))
```
