Implemente um código para avaliar o efeito do hiperparâmetro Núm_mín no algoritmo DBSCAN. Para tanto, considere ε=0,7 e os seguintes valores para o referido fator: 5, 10, 19, 20, 25.
```python
# Carrega as bibliotecas e funções necessárias
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Gera uma base de dados aleatória

X, y = make_blobs(n_samples=1000, centers=8, n_features=2, random_state=800)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
# Testa o algoritmo DBSCAN com ε = 0.7, e  os  seguintes  valores para 𝑁ú𝑚_𝑚í𝑛: 5, 10, 19, 20, 25

# Define uma lista de cores fixas para ruído (-1) e clusters (0, 1, 2, 3, 4, 5, 6, 7, 8)
# ruído = lightgray
# cluster 0 = red
# cluster 1 = blue
# cluster 2 = green
# cluster 3 = orange
# cluster 4 = purple
# cluster 5 = brown
# cluster 6 = pink
# cluster 7 = cyan
# cluster 8 = yellow
fixed_colors = ['lightgray', 'red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'cyan', 'yellow']

ep = 0.7
Num_min_values = [5, 10, 19, 20, 25]

for num_min in Num_min_values:
    db = DBSCAN(eps=ep, min_samples=num_min, metric='euclidean')
    labels = db.fit_predict(X)
    plot_colors = [fixed_colors[label + 1] if label != -1 else fixed_colors[0] for label in labels]
    plt.scatter(X[:,0], X[:,1], c=plot_colors)
    plt.title('Problema com diferentes épsilon (eps): ' + str(ep) + ' e Núm_min: ' + str(num_min))
    plt.show()
```
