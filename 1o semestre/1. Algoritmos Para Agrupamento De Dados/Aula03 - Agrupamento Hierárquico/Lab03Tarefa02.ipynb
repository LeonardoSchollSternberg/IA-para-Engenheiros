```python
# Importa as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import linkage
```

```python
# Gera os dados para o agrupamento
dados = pd.DataFrame({"X":[12,15,18,10,8,9,12,20], "Y":[6,16,17,8,7,6,9,18]})
dados
```

```python
# Plota o dendrograma pelo método de Ward
plt.title('Agrupamento Hierárquico: Dendograma')
dendrogram_plot = dendrogram(linkage(dados, method = 'ward'))
plt.xlabel('Ordem dos Dados')
plt.ylabel('Distância')
plt.show()
```

```python
# Aplica o método de Ward para o agrupamento hierárquico
num_clusters = 2
ac = AgglomerativeClustering(n_clusters = num_clusters, linkage='ward')
ac.fit(dados)
pred=ac.labels_
```

```python
# Plota as amostras com a identificação do cluster
plt.scatter(dados.X,dados.Y,c=pred, marker="o")
plt.title('Agrupamento Aglomerativo')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.show()
```
