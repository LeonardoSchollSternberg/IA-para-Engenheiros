```python
# Importa as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plts
%matplotlib inline
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import linkage
```

```python
# Gera os dados para o agrupamento
dados = pd.DataFrame({"X":[12,15,18,10,8,9,12,20], "Y":[6,16,17,8,7,6,9,18]})
dados
```

```python
# Aplica o método de Ward para o agrupamento hierárquico e plota o dendrograma
plt.title('Agrupamento Hierárquico: Dendograma')
dendrogram_plot = dendrogram(linkage(dados, method = 'ward'))
plt.xlabel('Ordem dos Dados')
plt.ylabel('Distância')
plt.show()
```

```python
# Aplica o método do vizinho mais próximo (single) para o agrupamento hierárquico e plota o dendrograma
plt.title('Agrupamento Hierárquico: Dendograma')
dendrogram_plot = dendrogram(linkage(dados, method = 'single'))
plt.xlabel('Ordem dos Dados')
plt.ylabel('Distância')
plt.show()
```

```python
# Aplica o método do vizinho mais distante (complete) para o agrupamento hierárquico e plota o dendrograma
plt.title('Agrupamento Hierárquico: Dendograma')
dendrogram_plot = dendrogram(linkage(dados, method = 'complete'))
plt.xlabel('Ordem dos Dados')
plt.ylabel('Distância')
plt.show()
```
