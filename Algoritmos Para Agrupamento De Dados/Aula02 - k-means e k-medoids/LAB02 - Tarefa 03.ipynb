Este programa trabalha com um tipo de gráfico denominado de blobs e foi implementado um cluster pelo algoritmo k-means. Vamos escolher o número de clusters, para esta base de dados, treinando o algoritmo k-means com diferentes valores de k e observando a “Soma dos Quadrados Dentro de cada Cluster” – WCSS. Basicamente a distância dos dados ao seu centroide é calculada e elevada ao quadrado. Em seguida, todas as distâncias ao quadrado são somadas para calcular o erro. Em função do valor do k ótimo obtido da Curva do Cotovelo treinar o modelo e avaliar o mesmo. Abrir o ambiente de programação Python, utilizar o código disponibilizado no arquivo e analisar o mesmo.

```python
# Importa as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
from sklearn.datasets import make_blobs

# criando dados
data, labels = make_blobs(n_samples=200, n_features=2, centers=5, cluster_std=1, random_state=150)
plt.scatter(data[:,0], data[:,1], c=labels,cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 150)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss, color='k')
plt.title('A Curva do Cotovelo do Algoritmo k-means')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Soma dos Quadrados do Cluster (wcss)')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, init='k-means++')
kmeans.fit(data)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))
ax1.set_title('K-means')
ax1.scatter(data[:,0],data[:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(data[:,0],data[:,1],c=labels,cmap='rainbow')
```
