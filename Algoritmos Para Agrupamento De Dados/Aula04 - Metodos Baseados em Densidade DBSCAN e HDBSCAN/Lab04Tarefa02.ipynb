```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Gera uma base de dados aleatória
X, y = make_blobs(n_samples=500, centers=4, n_features=2, random_state=800)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
# Testa o algoritmo DBSCAN com ε variando de 0.2, 0.7 e 4

# Define uma lista de cores fixas para ruído (-1) e clusters (0, 1, 2, 3)
# ruído = lightgray
# cluster 0 = red
# cluster 1 = blue
# cluster 2 = green
# cluster 3 = orange
fixed_colors = ['lightgray', 'red', 'blue', 'green', 'orange']

eps = [0.2,0.7,4]
for ep in eps:
    db = DBSCAN(eps=ep, min_samples=10, metric='euclidean')
    labels = db.fit_predict(X)
    plot_colors = [fixed_colors[label + 1] if label != -1 else fixed_colors[0] for label in labels]
    plt.scatter(X[:,0], X[:,1], c=plot_colors)
    plt.title('Problema com diferentes épsilon (eps): ' + str(ep))
    plt.show()
```
