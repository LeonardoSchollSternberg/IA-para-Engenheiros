Vamos implementar um agrupamento DBSCAN para comparar com modelos anteriormente estudados nesta disciplina. Métodos de agrupamento de partição, como k-means, e métodos de agrupamento hierárquico, como clustering aglomerativo, são bons para descobrir clusters esféricos ou convexos. Esses algoritmos são mais sensíveis a ruídos ou outliers, sendo assim, funcionam bem para clusters bem separados. Por outro lado, o DBSCAN é baseado na ideia principal de que cada dado de um grupo ou cluster tem um número mínimo de dados em um determinado raio. O principal objetivo do DBSCAN é descobrir a região densa que pode ser calculada usando o número mínimo de dados (𝑁ú𝑚_𝑚í𝑛) e um determinado raio (ε). O algoritmo DBSCAN tem a capacidade de gerar formas aleatórias de clusters e lidar com o ruído de um conjunto de dados. Também, não há necessidade de determinar o número de clusters, como por exemplo, no k-means. Portanto, o algoritmo DBSCAN identifica automaticamente o número de clusters de um conjunto de dados.
```python
# Importa as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_moons
import numpy as np
```

```python
# Gera uma base de dados
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
# Vamos implementar os métodos K-Means e hierárquico (Ward) para agrupar os dados em 2 clusters
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,3))
km = KMeans(n_clusters=2,random_state=0)
y_km = km.fit_predict(X)
ax1.scatter(X[y_km==0,0],X[y_km==0,1],c='lightblue',marker='o',s=40,label='cluster 1')
ax1.scatter(X[y_km==1,0],X[y_km==1,1],c='red',marker='s',s=40,label='cluster 2')
ax1.set_title('K-means clustering')

ac = AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
y_ac = ac.fit_predict(X)
ax2.scatter(X[y_ac==0,0],X[y_ac==0,1],c='lightblue',marker='o',s=40,label='cluster 1')
ax2.scatter(X[y_ac==1,0],X[y_ac==1,1],c='red',marker='s',s=40,label='cluster 2')
ax2.set_title('Agglomerative clustering')
plt.legend()
plt.show()
```

```python
# Aplicando o método DBSCAN
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.2,min_samples=5,metric='euclidean')
y_db = db.fit_predict(X)
plt.scatter(X[y_db==0,0],X[y_db==0,1],c='lightblue',marker='o',s=40,label='cluster 1')
plt.scatter(X[y_db==1,0],X[y_db==1,1],c='red',marker='s',s=40,label='cluster 2')
plt.legend()
plt.show()
```

```python
# Gera uma base de dados ruidosa
X, y = make_moons(n_samples=200, noise=0.10, random_state=0)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
# Vamos implementar os métodos hierárquico (Single) e o DBSCAN para agrupar os dados em 2 clusters

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,3))
db = DBSCAN(eps=0.2,min_samples=5,metric='euclidean')
y_db = db.fit_predict(X)
ax1.scatter(X[y_db==0,0],X[y_db==0,1],c='lightblue',marker='o',s=40,label='cluster 1')
ax1.scatter(X[y_db==1,0],X[y_db==1,1],c='red',marker='s',s=40,label='cluster 2')
ax1.set_title('DBSCAN clustering')

ac = AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='single')
y_ac = ac.fit_predict(X)
ax2.scatter(X[y_ac==0,0],X[y_ac==0,1],c='lightblue',marker='o',s=40,label='cluster 1')
ax2.scatter(X[y_ac==1,0],X[y_ac==1,1],c='red',marker='s',s=40,label='cluster 2')
ax2.set_title('Agglomerative clustering')
plt.legend()
plt.show()
```
