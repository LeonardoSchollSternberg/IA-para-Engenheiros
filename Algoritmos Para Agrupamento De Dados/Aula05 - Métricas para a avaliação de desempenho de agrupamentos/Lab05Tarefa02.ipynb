```python
# Importa as bibliotecas necessárias e a base de dados "diabetes"
import pandas as pd
```

```python
diabetes = pd.read_csv("./diabetes.csv")
diabetes.head()
```

```python
conj_caract = ['Pregnancies','Insulin','BMI','Age','Glucose','BloodPressure','DiabetesPedigreeFunction']
caracts = diabetes[conj_caract]
caracts = (caracts - caracts.min())/(caracts.max()-caracts.min())
alvo = diabetes.Outcome
caracts.head()
```

```python
# Aplica o método k-means para separar em dois agrupamentos e testa várias métricas de desempenho

from sklearn.model_selection import train_test_split
caract_trein, caract_teste, alvo_trein, alvo_teste = train_test_split(caracts, alvo,test_size=0.3, random_state=1)

from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import silhouette_score
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import fowlkes_mallows_score
num_clusters = 2
km = KMeans(n_clusters=num_clusters)
km.fit(caract_trein)
predicoes=km.predict(caract_teste)
print("Índice Davies-Bouldin:", davies_bouldin_score(caract_teste, predicoes))
print("Coeficiente da Silhueta:", silhouette_score(caract_teste, predicoes))
print("Adjusted Rand Score:", adjusted_rand_score(alvo_teste, predicoes))
print("F1-Score:", f1_score(alvo_teste, predicoes))
print("Escore Fowlkes Mallows:", fowlkes_mallows_score(alvo_teste, predicoes))
```
